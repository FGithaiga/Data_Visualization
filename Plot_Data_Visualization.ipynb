```python
#creat a sample dataset
#np.random.seed(10)#for reproducibility.

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.random.seed(10)
data={"Month":["Jan","Feb","Mar","Apr","May","Jun"],
     "Sales": np.random.randint(200, 500, 6),
     "Customers": np.random.randint(50, 200, 6),
     "Profit": np.random.randint(50, 150, 6),
}
np.random.seed(9)
data2={"Month":["Jan","Feb","Mar","Apr","May","Jun"],
     "Sales": np.random.randint(200, 500, 6),
     "Customers": np.random.randint(50, 200, 6),
     "Profit": np.random.randint(50, 150, 6),
}
data
df=pd.DataFrame(data)
df

data
df2=pd.DataFrame(data2)
df
```

```python
#plot a line graph(plot):Trend Over Time


plt.figure(figsize=(10, 7))
plt.plot(df['Month'], df['Sales'], color='b', marker='o', linestyle='-', label='Sales')
plt.plot(df2['Month'], df2['Sales'], color='g', marker='o', linestyle='-', label='Sales2')

plt.title('Sales vs Month') #Title of the plot
plt.xlabel('Month') #X-axis label
plt.ylabel('Sales') #Y-axis label
plt.legend() # shows the legend
plt.grid(True) #makes the grid visible(visual readable)
plt.show()
```

# X Axis:represents the bins.

## Boxplot:a visual that summarizes dat in 5 catergories.

#Minimum:is the lowest value(it excludes outliers).
#First Quartile;Q1:25th Percentile.
#Median;Q2:50th Percentile.
#Third Quartile:75th Percentile.
#Maximum:is the highest value(it excludes outliers).

## Scatter plot:a visual that shows the relationship between two variables.Each point represents an observation.

# X- shows one varriable
# Y- shows the other variable 
# bins number of values to be grouped together in a histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load an in build dataset

data=sns.load_dataset('tips')#inbuilt dataset
print(data)
```

```python
# Historgram

plt.figure(figsize=(10,7))
sns.histplot(data['total_bill'], bins=10, color='blue')
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
#Boxplot

plt.figure(figsize=(10,7))
sns.boxplot(x='day', y='tip', data=data, palette=['blue', 'green', 'yellow', 'red'])
plt.title('Tips Distribution per day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python

#scatter plot

plt.figure(figsize=(10,7))
sns.regplot(x='total_bill', y='tip', data=data, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'} )
plt.title('Tips Distribution Per Day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```
